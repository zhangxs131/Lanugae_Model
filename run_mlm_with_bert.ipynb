##使用pipeline函数进行测试

```python
from transformers import pipeline
from transformers import AutoTokenizer, BertForMaskedLM


text='If a girl really hate you ,she will [MASK] you to death.'

tokenizer = AutoTokenizer.from_pretrained("/content/t/checkpoint-579")
model = BertForMaskedLM.from_pretrained("/content/t/checkpoint-579")

generator = pipeline(task="fill-mask", model=model, tokenizer=tokenizer)

t=generator(text)
for i in t:
  print(i['score'])
  print(text.replace('[MASK]',i['token_str']))
```
